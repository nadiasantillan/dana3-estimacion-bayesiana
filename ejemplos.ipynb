```python
import preliz as pz
```

## Distrubuciones Continuas

### Uniforme

```python
pz.Uniform(0,1).plot_pdf()
pz.Uniform(2,6).plot_pdf()
```

### Normal

```python
pz.Normal(0,1).plot_pdf()
```

### Gamma

```python
pz.Gamma(mu=2, sigma=1).plot_pdf()
```

### Exponencial

```python
pz.Exponential(1).plot_pdf()
```

### Chi Cuadrado

```python
pz.ChiSquared(3).plot_pdf()
pz.ChiSquared(9).plot_pdf()
```

### Beta

```python
pz.Beta(4, 6).plot_pdf()
pz.Beta(15, 15).plot_pdf()

pz.Beta(1,1).plot_pdf()
```

### Weibull

```python
pz.Weibull(1, 2).plot_pdf()
```

### LogNormal

```python
pz.LogNormal(0, 1).plot_pdf()
```
